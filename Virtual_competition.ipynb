***DIABETES***

````python
df = pd.read_csv('/content/diabetes_model.csv')
df
````

Univariate Analysis

````python
# histograms
plt.hist(df['BloodPressure'],bins=10,color='red',edgecolor='black')
plt.xlabel('BloodPressure')
plt.ylabel('Range')
plt.title('Histogram of BloodPressure')
plt.show()
````

The histogram shows the distribution of blood pressure readings for a group of individuals. The x-axis represents the blood pressure readings, while the y-axis represents the frequency of each reading. The histogram has a bell-shaped curve, indicating that the majority of the readings are clustered around the middle of the distribution. The peak of the curve is around 70-80 mmHg

Bivariate Analysis

````python
import seaborn as sns
````

````python
# scatter plots
sns.scatterplot(x='BloodPressure',y='BMI',data=df)
plt.show()
````

there is a general trend for individuals with higher BMI to have higher blood pressure. However, individual variations exist, and other factors may also play a role in determining blood pressure

Multivariate Analysis

````python
# pairplot
sns.pairplot(df[['Age','Glucose','Insulin']])
plt.show()
````

Relationships between variables: The off-diagonal subplots show the relationships between pairs of variables. Some key observations include:
Negative correlation between age and glucose: As age increases, glucose levels tend to decrease.
Positive correlation between glucose and insulin: As glucose levels increase, insulin levels also tend to increase.
Weak correlation between age and insulin: There is a weak correlation between age and insulin, suggesting that the relationship is not very strong.
